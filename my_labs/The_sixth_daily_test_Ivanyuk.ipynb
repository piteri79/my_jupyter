### 1、 please read the flower photoes  and add them to a list.

```python
import os
import glob
from  skimage import io,transform
path = './flower_photos/' 
cate=[path+x for x in os.listdir(path) if os.path.isdir(path+x)]
images=[]
for i, folder in enumerate(cate):
    for im in glob.glob(folder+'/*.jpg'):
        img=io.imread(im)
        images.append(img)
print(type(images))
```

###  2、View the length of the previous generation list, take the last 11 elements to a new list.

```python
l = len(images)
print('длина первого списка = ', l)
images2 = []
for a in range(11):
    aa = images.pop(l - 1 - a)
    images2.append(aa)
print(type(images2),'\nдлина второго списка = ',len(images2))
```
